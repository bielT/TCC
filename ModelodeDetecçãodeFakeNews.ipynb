```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Dados de exemplo (substitua isso pelos seus próprios dados)
# Vamos criar uma série temporal simples para este exemplo.
data = np.array([i for i in range(100)])
target = np.array([i for i in range(1, 101)])

# Redimensionar os dados para (amostras, intervalo de tempo, características)
data = data.reshape((data.shape[0], 1, 1))

# Criar o modelo LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Treinar o modelo
model.fit(data, target, epochs=1000, validation_split=0.2, batch_size=1)

# Fazer previsões
test_input = np.array([i for i in range(100, 110)])
test_input = test_input.reshape((test_input.shape[0], 1, 1))
predictions = model.predict(test_input)

print(predictions)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Dados de exemplo (substitua isso pelos seus próprios dados)
texts = ["Este é um ótimo produto.", "Não gostei disso.", "Produto de alta qualidade.", "Não recomendo este produto."]
labels = [1, 0, 1, 0]  # 1 para positivo, 0 para negativo

# Tokenização e sequenciamento de texto
tokenizer = Tokenizer(num_words=1000)  # Número máximo de palavras a serem consideradas
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Padding para que todas as sequências tenham o mesmo comprimento
max_sequence_length = max([len(seq) for seq in sequences])
sequences = pad_sequences(sequences, maxlen=max_sequence_length)

# Criar o modelo LSTM
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=64, input_length=max_sequence_length))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compilar o modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Treinar o modelo
X = np.array(sequences)
y = np.array(labels)
model.fit(X, y, epochs=10, batch_size=2)

# Fazer previsões
new_texts = ["Adorei esse produto.", "Não poderia ser pior."]
new_sequences = tokenizer.texts_to_sequences(new_texts)
new_sequences = pad_sequences(new_sequences, maxlen=max_sequence_length)
predictions = model.predict(new_sequences)

for text, prediction in zip(new_texts, predictions):
    sentiment = "positivo" if prediction > 0.5 else "negativo"
    print(f"Texto: {text}, Sentimento: {sentiment}, Confiança: {prediction[0]:.2f}")
```
